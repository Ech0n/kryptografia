```python
def permute(k, perm):
    res = ""
    for idx in perm:
        res += k[idx]
    return res

def xor(bin_str1, bin_str2, keep_len = False):
    assert len(bin_str1) == len(bin_str2)
    initLen = len(bin_str1)
    res = ""
    for i in range(len(bin_str1)):
        res += str(int(bin_str1[i]) ^ int(bin_str2[i]))
    if keep_len:
        while len(res)<initLen:
            res = "0"+res
    return res

def Feistel(message, subkeys, F):
    assert len(message) == 64
    la = list(message[:32])
    r = list(message[32:])
    for subkey in subkeys:
        la = xor(F(r, subkey), la)
        r, la = la, r
    key = str(r) + str(
        la
    )  ##Dodaje prawy do lewego, bo w ostatnije iteracji nie ma zamiany r = l, l = r
    return key


def bin2dec(bin_str):
    assert len(bin_str) > 0 and all([digit in ["0", "1"] for digit in bin_str])
    num = 0
    for c in bin_str:
        num = (num << 1) + int(c)
    return num

def dec2bin(dec, pad):
    binary = ""
    for i in range(pad):
        prefix = "0"
        if dec > 0:
            prefix = str(dec % 2)
            dec //= 2
        binary = prefix + binary
    return binary



def F(right, subkey):
    e = permute(right, E)
    x = xor(e, subkey)
    key = ""
    for i in range(8):
        idx = i * 6
        group = x[idx : idx + 6]
        row = bin2dec(group[0] + group[5])
        col = bin2dec(group[1:5])
        s = SBox[i][row][col]
        key += dec2bin(s, 4)

    key = permute(key, P)
    return key
def DES(message, subkeys):
    key = permute(message, IP)
    key = Feistel(key, subkeys, F)
    key = permute(key, FP)
    return key
```

```python

def shift_left(tab, n):
    arr = np.zeros_like(tab)
    k = len(tab)
    for num in range(k):
        arr[(num - n) % k] = tab[num]
    return arr
def key_schedule(key):
    assert len(key) == 64
    assert len(shift_table) == 16
    subkeys = []
    k = permute(key, PC1)
    l = list(k[:28])
    r = list(k[28:])
    for shift in shift_table:
        l = shift_left(l, shift)
        r = shift_left(r, shift)
        key_sum = "".join(l) + "".join(r)
        new_key = permute(key_sum, PC2)
        subkeys.append(new_key)
    return subkeys
```

```python
def CBC(msg,key,iv):
    subkeys = key_schedule(key)
    enc = []
    blocks = image_split(msg, 64)
    carry = iv
    for block in blocks:
        x = xor(block, carry, keep_len=True)
        y = DES(x, subkeys)
        enc.append(y)
        carry = y
    return "".join(enc)
```

Zaimplementuj deszyfrację w trybi CBC. Zdeszyfruj zaszyfrowany w poprzednim zadaniu obrazek i sprawdź czy otrzymałeś pierwotny obrazek.

<img src="CBC_decryption.png" width="1000" align="left"/>

```python
def CBCde(msg,key,iv):
    subkeys = key_schedule(key)[::-1]
    enc = []
    blocks = image_split(msg, 64)
    carry = iv
    for block in blocks:
        y = DES(block, subkeys)
        enc.append(xor(carry,y))
        carry = block
    return "".join(enc)
```

```python
def OFB(msg,key,iv):
    subkeys = key_schedule(key)
    enc = []
    blocks = image_split(msg, 64)
    carry = iv
    for block in blocks:
        y = DES(carry, subkeys)
        enc.append(xor(block,y))
        carry = y
    return "".join(enc)
```

## Zadanie 5 (1 pkt.)

Zaimmplementuj funkcję dodającą do wiadomości padding zgodnie z standardem PKCS#5. Wiadomości będą zapisywane w postaci tablicy bajtów zapisanych szesnastkowo. Funkcja ma przyjmować jako argumenty wiadomość oraz długość bloku do jakiego wiadomość ma być dopełniona.

```python
def padding(msg, L):
    missing = L - (len(msg) % L)
    if missing == 0:
        missing = L

    hexa = hex(missing)[2:]
    while len(hexa) < 2:
        hexa = "0"+hexa
    return msg + [hexa] * missing
```

## Zadanie 6 (2 pkt.)

Zaimplementuj wyrocznię, która sprawdza czy podana wiadomość ma prawidłowy padding i zwraca **False** lub **True**.

```python
def oracle(msg, L):
    if len(msg) == 0:
        return False

    last = msg[-1]

    pad = int(last, 16)
    if pad > L:
        return False

    if len(msg) < pad:
        return False

    pad_hex = hex(pad)[2:]
    while len(pad_hex) < 2:
        pad_hex = "0" + pad_hex

    for b in msg[-pad:]:
        if b != pad_hex:
            return False

    return True
```
